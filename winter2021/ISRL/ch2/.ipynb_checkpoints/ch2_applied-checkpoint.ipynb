##### c.3 Produce a side-by-side boxplot of Outstate VS Private

```python
college.boxplot(column=['Outstate'], by='Private')
```

A boxplot is a method to depict groups of numerical data using their quartiles. I has a **5 number summary**: minimum, Q1, median, Q3 and maximum. In this case, the **Outstate** variable describes the out of state tuition. The **Private** variable shows if the university is Public or Private. In the right hand side (YES column), the box plot suggests that for Private universities, the median tuition (or 50th) for an out-of-state student is around 11,000, whereas in Public universities, the median out-of-state is close to 7,000. This figures reflects that out-of-state students pay a considerable higher tuition in Private universities than in Public ones.
